```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('loan_data_2007_2014.csv')
```

```python
default_loans = ['Charged Off', 'Default', 'Does not meet the credit policy. Status:Charged Off', 'Late (31-120 days)']

data['good_bad'] = np.where(data['loan_status'].isin(default_loans), 0, 1)
```

```python
inputs_train, input_test, targets_train, targets_test = train_test_split(data.drop('good_bad', axis=1),
                                                                          data['good_bad'],
                                                                          train_size = 0.25,
                                                                          random_state = 102)
```

```python
data_acutal = pd.concat([inputs_train, targets_train], axis=1)
```

```python
data_acutal.head()
```

```python
data_acutal.reset_index(inplace=True, drop=True)
```

```python
data_acutal.to_csv('new_data.csv')
```
